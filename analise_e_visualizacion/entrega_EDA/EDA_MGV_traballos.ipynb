# 2. Traballo no dataframe

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(url2)
df.info()
```

Para traballar mellor no dataframe, recheamos datos e eliminamos malos datos

```python
try:
    df['Age'] = df['Age'].fillna(df.Age.median())
    df['Gender'] = df['Gender'].fillna('NS/NC')
    df['City'] = df['City'].fillna('NS/NC')
    df['Position'] = df['Position'].fillna('NS/NC')
    df['Experience'] = df['Experience'].fillna(0)
    df['Brutto'] = df['Brutto'].fillna(df.Brutto.median())
    df['Language@Work'] = df['Language@Work'].fillna('NS/NC')
    df['Company size'] = df['Company size'].fillna('NS/NC')
    df['Company type'] = df['Company type'].fillna('NS/NC')
    
    df.rename(columns={'Seniority level':'Seniority',
                       'Сontract duration': 'Duration'}, inplace=True)
    df.drop(['Company business sector','Employment status'],axis=1,inplace=True)
 
    df[df['Experience'] == '1 (as QA Engineer) / 11 in total'] = '11'
    df[df['Experience'] == '2,5'] = '2.5'
    df[df['Experience'] == '15, thereof 8 as CTO'] = '15'
    df[df['Experience'] == '6 (not as a data scientist, but as a lab scientist)'] = '6'
    df[df['Experience'] == '383'] = '4'
    df[df['Experience'] == 'less than year'] = '0.5'
    df[df['Experience'] == '1,5'] = '1.5'
    df['Experience'] = df['Experience'].astype(float)
    df['Brutto'] = df['Brutto'].astype(float)
    df[df['Brutto'] > 500000.0] = 100000.
    df[df['Experience'] > 50] = 50
    df['Age'] = df['Age'].astype(float)
except:
    print('No funcionó')
   
```
